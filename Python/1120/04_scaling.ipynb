# 서로 다른 범위의 데이터를 같은 범위로 스케일링

```python
import pandas as pd
import numpy as np
```

### 1. 데이터

```python
df = pd.DataFrame({
    'A': [14.00, 90.20, 90.95, 96.27, 91.21],
    'B': [103.02, 107.26, 110.35, 114.23, 114.68],
    'C': ['big', 'small', 'big', 'small', 'small']
})

display(df)
```

### 2. 최소값 - 최대값 정규화

공식 : (Xi - min(X)) / (max(X) - min(X))

```python
(df["A"] - df["A"].min()) / (df["A"].max() - df["A"].min())
```

### 3. Z-Score 정규화

공식 : (Xi - mean(X)) / std(X)

```python
(df["B"] - df["B"].mean()) / df["B"].std()
```
