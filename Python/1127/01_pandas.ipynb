### 2-1. 평균값으로 처리

```python
df = pd.DataFrame(np.random.randn(8,3),
				  columns=["C1","C2","C3"])

df.loc[2, "C2"] = np.nan
df.loc[3, "C2"] = np.nan
display(df)

display(df.mean())

df.fillna(df.mean())
```

## 3. 이상치 제거

```python
df = pd.DataFrame(np.random.randn(8,3),
				  columns=["C1","C2","C3"])

df.loc[1, "C1"] = 11
df.loc[3, "C2"] = -10

plt.boxplot([df["C1"], df["C3"]])
plt.show()
```
